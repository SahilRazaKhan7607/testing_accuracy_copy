```python

import pandas as pd
from sklearn.naive_bayes import GaussianNB

# 1. Load your training data
data = pd.read_excel('iris_dataset.xlsx')
X = data.iloc[:, :-1]  # Features (Measurements)
y = data.iloc[:, -1]   # Label (Flower Name)

# 2. Create and train the model
model = GaussianNB()
model.fit(X, y)

# 3. Load your NEW testing data
datatest = pd.read_excel("iris_testing.xlsx")

# 4. Prepare the test features (must be the same columns as X)
x_test = datatest.iloc[:, :-1] 

# 5. Make predictions
predictions = model.predict(x_test)

# 6. See the results
print(predictions)
```

Testing accuracy of model


```python
y_test=datatest.iloc[: ,-1]
actual=y_test.values
count=0
for i in range(0,len(actual)):
  if actual[i]==predictions[i]:
    count=count+1
print(count*100/len(actual))   
```
